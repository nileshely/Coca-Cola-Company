# Coca-Cola Company Market Performance (2019-2024)

The dataset provides a comprehensive overview of the market performance of the Coca-Cola Company from 2019 to 2024, offering daily insights into stock prices, trading volume, and market trends. Analysis of the data reveals intriguing patterns and trends that shed light on the company's financial health and investor sentiment.

Firstly, examining the daily closing prices of Coca-Cola's stock unveils the company's trajectory over the five-year period. A line plot illustrates fluctuations in stock prices, highlighting periods of growth, decline, and stability. From the visualizations, investors can identify key turning points and assess the effectiveness of strategic initiatives undertaken by the company during this period.

Secondly, analyzing trading volume provides valuable insights into investor activity and market sentiment. A histogram showcases the distribution of trading volume, indicating periods of heightened activity or indifference among investors. By correlating trading volume with price movements, analysts can gauge the level of market participation and the degree of consensus regarding the company's prospects.

Thirdly, candlestick charts offer a detailed depiction of daily price movements, including opening, closing, high, and low prices. These visualizations provide a holistic view of market dynamics, allowing analysts to identify patterns such as trends, reversals, and volatility. By interpreting candlestick patterns, investors can make informed decisions regarding entry and exit points in the market.

Fourthly, box plots of high and low prices offer a comparative analysis of price ranges, highlighting the variability and distribution of prices over time. This visualization aids in understanding the level of price dispersion and the potential for price extremes. By assessing the spread of high and low prices, investors can gauge the level of market uncertainty and adjust their risk management strategies accordingly.

Lastly, scatter plots and density plots provide additional insights into the relationship between different variables, such as high and low prices or the distribution of closing prices. These visualizations facilitate a deeper understanding of market dynamics and serve as valuable tools for identifying correlations and outliers. By leveraging these insights, investors can refine their trading strategies and optimize their investment decisions.

In conclusion, the dataset on Coca-Cola Company's market performance from 2019 to 2024 offers a wealth of information for investors, analysts, and stakeholders. Through the use of various data visualizations, including line plots, histograms, candlestick charts, box plots, scatter plots, and density plots, market trends, volatility, and investor sentiment can be effectively analyzed and interpreted. These insights empower market participants to make informed decisions and navigate the dynamic landscape of the stock market with confidence.

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/kaggle/input/coca-cola-company/Coca Cola.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

plt.figure(figsize=(10,6))
plt.plot(data['Date'], data['Close'], color='red')
plt.title('Coca-Cola Company - Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(data['Volume'], bins=30, color='darkred', alpha=0.7)
plt.title('Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from mplfinance.original_flavor import candlestick_ohlc
import matplotlib.dates as mdates

# Convert 'Date' to mdates format
data['Date'] = data['Date'].apply(mdates.date2num)

# Plot
plt.figure(figsize=(12,8))
ax = plt.subplot()
candlestick_ohlc(ax, data.values, width=0.6, colorup='red', colordown='red')
ax.xaxis_date()
plt.title('Coca-Cola Company - Candlestick Chart')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([data['High'], data['Low']], labels=['High', 'Low'], patch_artist=True, boxprops=dict(facecolor='red', color='red'))
plt.title('High and Low Prices Box Plot')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['High'], data['Low'], color='darkred', alpha=0.5)
plt.title('High vs. Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
data['Close'].plot(kind='density', color='red')
plt.title('Density Plot of Close Prices')
plt.xlabel('Close Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
